```python
import  pandas as pd
import numpy as np

# creating matrix
M1 = np.random.randint (-10, 11, size=(3,3))
M2 = np.random.randint (-10, 11, size=(3,3))
M3 = np.random.randint (-10, 11, size=(3,3))

print(M1)
print(M2)
print(M3)

# inverses a matrix
M1_inverse = np.linalg.inv(M1)
M2_inverse = np.linalg.inv(M2)
M3_inverse = np.linalg.inv(M3)

print(M1_inverse)
print(M2_inverse)
print(M3_inverse)

# computes the product of a matrix
product1 = M1 @ M1_inverse
product2 = M2 @ M2_inverse
product3 = M3 @ M3_inverse

print(product1)
print(product2)
print(product3)

```

```python
df = pd.read_csv('kc_house_data.csv')

# drops columns id, date, and zipcode
df = df.drop(columns=['id', 'date', 'zipcode'])
df.head

summary_stats = pd.DataFrame({
    "average": df.mean(),
    "min": df.min(),
    "max": df.max(),
    "variance": df.var()
})

print(summary_stats)

lowest_avg = summary_stats["average"].idxmin()
highest_avg = summary_stats["average"].idxmax()

print(f"Lowest Average: {lowest_avg} = {summary_stats.loc[lowest_avg, 'average']:.2f}")
print(f"Highest Average: {highest_avg} = {summary_stats.loc[highest_avg, 'average']:.2f}")

lowest_var = summary_stats["variance"].idxmin()
highest_var = summary_stats["variance"].idxmax()

print(f"Lowest Variance: {lowest_var} = {summary_stats.loc[lowest_var, 'variance']:.2f}")
print(f"Highest Variance: {highest_var} = {summary_stats.loc[highest_var, 'variance']:.2f}")
```
